## Questão 1

Dado 2 v.a. $X$ e $Y$ Normais, independentes com $\mu = 0$ e $\sigma^2 = 1$, gere sequências de 5.000 amostras cada para uma delas.

### Covariância

$$
 \text{Cov}(X,Y) = \mathbb{E} \left[ (X - \mathbb{E}[X])(Y - \mathbb{E}[Y])^\top \right]
$$

<br><br>

$$
\text{Cov}(X,Y) = \frac{1}{n} \sum_{i=1}^{n} (X_i - \bar{X})(Y_i - \bar{Y})^\top
$$

<br><br>

$$
\Sigma_{XY} =
\begin{bmatrix}
\text{Var}(X) & \text{Cov}(X,Y) \\
\text{Cov}(Y,X) & \text{Var}(Y)
\end{bmatrix}
$$

<br>

Se as variáveis aleatórias X e Y forem independentes, então:

$$
 \text{Cov}(X,Y) = 0
$$


### Correlação

$$
\rho(X,Y) = \dfrac{Cov(X,Y)}{\sigma_X\sigma_Y}
$$

#### Função auxiliar

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

def analisar_variaveis(rv1, rv2, nome1="X", nome2="Y", limits=None):
    # Matriz de correlação
    corr = np.corrcoef(rv1, rv2)
    fig, ax = plt.subplots(figsize=(6, 5))
    sns.heatmap(corr, annot=True, xticklabels=[nome1, nome2], yticklabels=[nome1, nome2], ax=ax)
    plt.title(f"Matriz de Correlação entre {nome1} e {nome2}")
    plt.show()

    # Matriz de covariância
    cov = np.cov(rv1, rv2)
    fig, ax = plt.subplots(figsize=(6, 5))
    sns.heatmap(cov, annot=True, xticklabels=[nome1, nome2], yticklabels=[nome1, nome2], ax=ax)
    plt.title(f"Matriz de Covariância entre {nome1} e {nome2}")
    plt.show()

    # Gráfico de dispersão
    plt.figure(figsize=(7, 6))
    plt.scatter(rv1, rv2, alpha=0.3)
    plt.xlabel(nome1)
    plt.ylabel(nome2)
    plt.title(f"Dispersão de {nome1} vs {nome2}")
    if limits:
        plt.xlim(limits)
        plt.ylim(limits)
    plt.show()
```

## Questão 02

Considere uma função densidade de probabilidade (fdp) Normal bidimensional $f_{X,Y}(X,Y)$, com médias $m_x = 5$, $m_y= 3$ e matriz de covariância $P = [4 1.8;1.8 1]$. Gere 100.000 amostras para essa distribuição.

A função de densidade de probabilidade (fdp) da distribuição Normal bidimensional é dada por:


$f_{X,Y}(x, y) = \frac{1}{2\pi \sigma_x \sigma_y \sqrt{1 - \rho^2}} \exp\left( -\frac{1}{2(1 - \rho^2)} \left[ \left( \frac{x - \mu_x}{\sigma_x} \right)^2 - 2\rho \left( \frac{x - \mu_x}{\sigma_x} \right) \left( \frac{y - \mu_y}{\sigma_y} \right) + \left( \frac{y - \mu_y}{\sigma_y} \right)^2 \right] \right)$


Encontrando os valores dos parâmetros:

- Médias:

  - $\mu_x = 5$

  - $\mu_y = 3$

- Desvio Padrão:

  - $\sigma_x = \sqrt{4} = 2$

  - $\sigma_y = \sqrt{1} = 1$

- Matriz de covariância:

$$
P = \begin{bmatrix}
\sigma_x^2 & \sigma_{xy} \\
\sigma_{xy} & \sigma_y^2
\end{bmatrix} = \begin{bmatrix}
4 & 1.8 \\
1.8 & 1
\end{bmatrix} \Rightarrow \sigma_{x,y} = 1.8
$$

- Coeficiente de correlação entre $X$ e $Y$, dado por $\rho = \dfrac{\sigma_{xy}}{\sigma_x \sigma_y} = \dfrac{1.8}{2 \cdot 1} = 0.9$.

---

Substituindo os valores dos parâmetros:

- $\mu_x = 5$
- $\mu_y = 3$
- $\sigma_x = 2$
- $\sigma_y = 1$
- $\rho = 0.9$

temos:

$$
f_{X,Y}(x, y) = \frac{1}{2\pi (2)(1) \sqrt{1 - (0.9)^2}}
\exp\left( -\frac{1}{2(1 - 0.9^2)}
\left[ \left( \frac{x - 5}{2} \right)^2
- 2(0.9) \left( \frac{x - 5}{2} \right)\left( y - 3 \right)
+ (y - 3)^2 \right] \right)
$$

### Simplificando a constante:

- $2\pi (2)(1) = 4\pi$
- $1 - 0.9^2 = 1 - 0.81 = 0.19$

$$
\frac{1}{2\pi \sigma_x \sigma_y \sqrt{1 - \rho^2}}
= \frac{1}{4\pi \sqrt{0.19}}
$$

### Simplificando o expoente:

- $2(1 - 0.9^2) = 2(0.19) = 0.38$

$$
-\frac{1}{2(1-\rho^2)} = -\frac{1}{0.38}
$$

Portanto, o termo interno fica:

$$
\left( \frac{x - 5}{2} \right)^2 - 1.8\left(\frac{x - 5}{2}\right)(y - 3) + (y - 3)^2
$$

### Resultado final:

$$
f_{X,Y}(x,y) = \frac{1}{4\pi\sqrt{0.19}} \,
\exp\left( -\frac{1}{0.38}\left[ \left(\frac{x - 5}{2}\right)^2 - 1.8\left(\frac{x - 5}{2}\right)(y - 3) + (y - 3)^2 \right] \right)
$$


Esta expressão fornece a probabilidade conjunta das variáveis $X$ e $Y$ assumirem valores específicos, considerando suas médias, variâncias e a correlação entre elas.



### a) Plote o Gráfico 3D de $f_{X,Y}(x, y)$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal
from mpl_toolkits.mplot3d import Axes3D

def plota_fdp_bidimensional(mean, cov):
    # Criar uma grade de valores (±3 desvios-padrão)
    x = np.linspace(mean[0] - 3*np.sqrt(cov[0][0]), mean[0] + 3*np.sqrt(cov[0][0]), 100)
    y = np.linspace(mean[1] - 3*np.sqrt(cov[1][1]), mean[1] + 3*np.sqrt(cov[1][1]), 100)
    X_grid, Y_grid = np.meshgrid(x, y)

    # Avaliar a fdp da Normal bivariada em cada ponto da grade
    pos = np.dstack((X_grid, Y_grid))
    rv = multivariate_normal(mean, cov)
    Z = rv.pdf(pos)

    # Plotar o gráfico 3D
    fig = plt.figure(figsize=(10, 7))
    ax = fig.add_subplot(111, projection='3d')
    ax.plot_surface(X_grid, Y_grid, Z, cmap='viridis')
    ax.set_title('Gráfico 3D da fdp Normal Bidimensional')
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('f_XY(X,Y)')
    plt.show()

    return X_grid, Y_grid, Z
```

```python
# Parâmetros da distribuição
mean = [5, 3]
cov = [[4, 1.8],
       [1.8, 1]]

X_grid, Y_grid, Z = plota_fdp_bidimensional(mean, cov)
```

### b) Plote as superfícies de níveis de $f_{X,Y}(x, y)$

Para o item (b), vamos plotar as curvas de nível (contornos) de Fx,y(X,Y)

```python
# Plotar as curvas de nível (contornos)
plt.figure(figsize=(8, 6))
plt.contour(X_grid, Y_grid, Z, levels=10, cmap='viridis')
plt.title('Curvas de Nível da fdp Normal Bidimensional')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```
